```python
import pandas as pd
```

# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("cumulative.csv")
df = df.drop(columns=["rowid", "kepid", "kepoi_name", "kepler_name", "koi_pdisposition", "koi_score", "koi_tce_delivname"])
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
target = df["koi_disposition"]
#target.unique()
target_names = ["CONFIRMED", "FALSE POSITVE", "CANDIDATE"]
```

```python
data = df.drop("koi_disposition", axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=43)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

# Train the Support Vector Machine

```python
from sklearn.svm import SVC 
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
model.score(X_train_scaled, y_train)
```

```python
model2 = SVC(kernel='linear')
model2.fit(X_test_scaled, y_test)
```

```python
model2.score(X_test_scaled, y_test)
```

```python
print(f"Training Data Score: {model2.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {model2.score(X_test_scaled, y_test)}")
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the `C` and `gamma` parameters

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [1, 5, 10], 'gamma': [0.0001, 0.001]}
grid = GridSearchCV(model, param_grid=param_grid)
```

```python
# Train the model with GridSearch
grid.fit(X_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
